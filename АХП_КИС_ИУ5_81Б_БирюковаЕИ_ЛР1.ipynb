### **1. Загрузка данных**

```r
# Загрузка необходимых библиотек
if (!require(readxl)) install.packages("readxl", repos = "https://cloud.r-project.org")
if (!require(dplyr)) install.packages("dplyr", repos = "https://cloud.r-project.org")
if (!require(ggplot2)) install.packages("ggplot2", repos = "https://cloud.r-project.org")
if (!require(tidyr)) install.packages("tidyr", repos = "https://cloud.r-project.org")

library(readxl)
library(dplyr)
library(ggplot2)
library(tidyr)

# Загрузка данных из CSV файла
Data <- read.csv("https://raw.githubusercontent.com/junaart/AHP_KIS/refs/heads/main/Lab_1/computer_dataset_1000.csv", sep = ",")
View(Data)
colnames(Data)
cat("Размер датасета:", nrow(Data), "x", ncol(Data), "\n")

# Приводим значения скорости Ethernet к числовым значениям в Gbps
ethernet_speed_value <- as.numeric(gsub("[^0-9.]", "", Data$Ethernet_Speed))

is_gbps <- grepl("gbps", Data$Ethernet_Speed, ignore.case = TRUE)
is_mbps <- grepl("mbps", Data$Ethernet_Speed, ignore.case = TRUE)

# Если явно Gbps — оставляем как есть
# Если явно Mbps или единицы не указаны — считаем что это Mbps и переводим в Gbps
Data$Ethernet_Speed_Gbps <- ifelse(is_gbps, ethernet_speed_value,
                            ifelse(is_mbps, ethernet_speed_value / 1000,
                                   ethernet_speed_value / 1000))

Data$Ethernet_Speed_Gbps <- as.numeric(Data$Ethernet_Speed_Gbps)
summary(Data$Ethernet_Speed_Gbps)
```

### **2. Фильтрация по требованиям**

```r
# Отбор по критериям варианта 5 (Криптовалютная биржа, HFT)
C <- Data$ID[
  (Data$CPU_Cores >= 16) &
  (Data$RAM_GB >= 64) &
  (Data$Storage_GB >= 1000) &
  (Data$Ethernet_Speed_Gbps  >= 2.5) &
  (Data$Price_USD <= 8000) &
  (Data$Storage_Type %in% c("SSD", "NVMe")) &
  (Data$Condition %in% c("Like New", "New")) &
  (!is.na(Data$Power_Consumption_W)) &
  (Data$Power_Consumption_W <= 800)
]

C
```

```r
# Убираем пропущенные значения
C <- C[!is.na(C)]
cat("Количество конфигураций, прошедших отбор:", length(C), "\n")
C
```

### **3. Парето-оптимизация**

```r
# Формирование дата фрейма для парето оптимизации
F <- data.frame(
  Data$CPU_Cores[Data$ID %in% C],
  Data$RAM_GB[Data$ID %in% C],
  Data$Storage_GB[Data$ID %in% C],
  Data$Ethernet_Speed_Gbps[Data$ID %in% C],
  1 / Data$Price_USD[Data$ID %in% C],
  1 / Data$Power_Consumption_W[Data$ID %in% C]
)

colnames(F) <- c(
  "CPU_Cores",
  "RAM_GB",
  "Storage_GB",
  "Ethernet_Speed_Gbps",
  "Inv_Price_USD",
  "Inv_Power_W"
)

rownames(F) <- C

View(F)
```

```r
# Проверка: X доминирует по Парето Y (X >_p Y)

# Для отобранных 6 конфигураций строится матрица критериев:
# Максимизируемые: CPU_Cores, RAM_GB, Storage_GB, Ethernet_Speed_Gbps
# Минимизируемые: цена и энергопотребление (берутся обратные величины)
# Алгоритм: проверяется доминирование по Парето — конфигурация X доминирует Y, если X не хуже по всем критериям и лучше хотя бы по одному.

DPareto <- function(X, Y) {
  p <- TRUE
  l <- FALSE
  i <- 1
  while (p & (i <= length(X))) {
    if (X[i] < Y[i]) p <- FALSE
    if (X[i] > Y[i]) l <- TRUE
    i <- i + 1
  }
  if (!p | !l) return(FALSE)
  else return(TRUE)
}

result <- c()
for (i in c(1:length(rownames(F)))) {
  p <- TRUE
  for (j in c(1:length(rownames(F)))) {
    if (DPareto(F[j, ], F[i, ]))
      p <- FALSE
  }
  if (p) result <- c(result, rownames(F)[i])
}
cat("Парето-оптимальные конфигурации:", length(result), "\n")
result
```

```r
FF <- data.frame(
  Data$CPU_Cores[Data$ID %in% result],
  Data$RAM_GB[Data$ID %in% result],
  Data$Storage_GB[Data$ID %in% result],
  Data$Ethernet_Speed_Gbps[Data$ID %in% result],
  Data$Price_USD[Data$ID %in% result],
  Data$Power_Consumption_W[Data$ID %in% result]
)

colnames(FF) <- c(
  "CPU_Cores",
  "RAM_GB",
  "Storage_GB",
  "Ethernet_Speed_Gbps",
  "Price_USD",
  "Power_Consumption_W"
)

rownames(FF) <- result
FF
```

### **4. Нормирование критериев**

```r
# Каждый критерий нормируется в диапазон [0,100]:
# Для максимизируемых: (X - min) * 100 / (max - min)
# Для минимизируемых: (max - X) * 100 / (max - min)

# CPU
mmax <- max(FF$CPU_Cores)
mmin <- min(FF$CPU_Cores)
CPU_Cores_norm <- (FF$CPU_Cores - mmin) * 100 / (mmax - mmin)

# RAM
mmax <- max(FF$RAM_GB)
mmin <- min(FF$RAM_GB)
RAM_GB_norm <- (FF$RAM_GB - mmin) * 100 / (mmax - mmin)

# Storage
mmax <- max(FF$Storage_GB)
mmin <- min(FF$Storage_GB)
Storage_GB_norm <- (FF$Storage_GB - mmin) * 100 / (mmax - mmin)

# Network (чем больше — тем лучше)
mmax <- max(FF$Ethernet_Speed_Gbps)
mmin <- min(FF$Ethernet_Speed_Gbps)
Ethernet_Speed_Gbps_norm <- (FF$Ethernet_Speed_Gbps - mmin) * 100 / (mmax - mmin)

# Цена (чем меньше — тем лучше)
mmax <- max(FF$Price_USD)
mmin <- min(FF$Price_USD)
Price_USD_norm <- (mmax - FF$Price_USD) * 100 / (mmax - mmin)

# Энергопотребление (чем меньше — тем лучше)
mmax <- max(FF$Power_Consumption_W)
mmin <- min(FF$Power_Consumption_W)
Power_Consumption_W_norm <- (mmax - FF$Power_Consumption_W) * 100 / (mmax - mmin)

WW <- data.frame(
  CPU_Cores_norm,
  RAM_GB_norm,
  Storage_GB_norm,
  Ethernet_Speed_Gbps_norm,
  Price_USD_norm,
  Power_Consumption_W_norm
)

rownames(WW) <- rownames(FF)

View(WW)
```

### **5. Идеальная точка и расстояние**

```r
# Идеальная точка — это наилучшие значения по каждому критерию: (100, 100, 100, 100, 100, 100)
# Вычисляется евклидово расстояние от каждой конфигурации до идеальной точки

distance <- function(A, B) {
  return(sqrt(sum((A - B)^2)))
}

distances <- sapply(result, function(i) distance(ideal, WW[i, ]))
names(distances) <- result
for (i in result) {
  print(paste(i, ":", as.character(distance(ideal, WW[i, ]))))
}
```

```r
# Конфигурация, ближайшая к идеальной точке
best_id <- result[which.min(distances)]
best_distance <- min(distances)
cat("Рекомендуемая конфигурация (по расстоянию до идеальной точки):", best_id, "\n")
cat("Расстояние до идеальной точки:", round(best_distance, 4), "\n\n")
Data[Data$ID == best_id, ]
```

### **6. Линейная свертка с приоритетами**

```r
# Задаются веса критериев: a = (3, 3, 2, 3, 2, 3) — что соответствует приоритетам:
# CPU (высокий)
# RAM (высокий)
# Storage (средний)
# Network (высокий)
# Price (средний)
# Power (высокий)
# Веса нормализуются: anorm = a / sum(a)

# CPU, RAM, Storage, Network, Price, Power
a <- c(3, 3, 2, 3, 2, 3)
anorm <- a / sum(a)
print(anorm)
sum(anorm)

scores <- sapply(result, function(i) sum(anorm * WW[i, ]))
names(scores) <- result
for (i in result) {
  print(paste(i, ":", as.character(sum(anorm * WW[i, ]))))
}
```

```r
# Лучшая конфигурация по взвешенной свертке
best_weighted_id <- result[which.max(scores)]
best_weighted_score <- max(scores)
cat("\nРекомендуемая конфигурация (по линейной свертке):", best_weighted_id, "\n")
cat("Счет:", round(best_weighted_score, 4), "\n\n")
Data[Data$ID == best_weighted_id, ]
```

### **7. Определение весовых коэффициентов для сегментов**

Требуется определить оптимальное распределение ресурсов между сегментами инфраструктуры **криптовалютной биржи (HFT)** с учётом требований к низкой задержке, высокой пропускной способности сети и надёжности хранения:

1. **Trading сегмент (6–10 узлов)** — matching engine, обработка ордеров, критична минимальная задержка
2. **Gateways сегмент (3–5 узлов)** — приём и маршрутизация трафика, сетевое ускорение, RDMA
3. **DB сегмент (3–5 узлов)** — хранение состояния и журналирование, высокая надёжность, быстрый NVMe

**Требуемый масштаб:**

| Параметр              | Trading (6–10)            | Gateways (3–5)            | DB (3–5)                  | Суммарно        |
|:----------------------|:--------------------------|:--------------------------|:--------------------------|:----------------|
| Количество узлов      | 6–10                      | 3–5                       | 3–5                       | **12–20**       |
| CPU (ядер на узел)    | 16–24                     | 16–24                     | 24–32                     | —               |
| RAM (на узел)         | 64–128 ГБ                 | 64–128 ГБ                 | 128–256 ГБ                | —               |
| Storage (на узел)     | 0.5–1 ТБ **Optane**        | 0.5–1 ТБ **NVMe**          | 2–4 ТБ **NVMe**            | —               |
| Сеть                  | **10–25G RDMA**            | **10–25G RDMA**            | **10G**                   | —               |
| Цена (за узел)        | `$15–25K`                 | `$12–18K`                 | `$18–30K`                 | —               |

Далее рассчитываются доли ресурсов (CPU/RAM/Storage) между сегментами и формируются приоритеты распределения, чтобы обеспечить баланс производительности, стоимости и энергопотребления при масштабировании инфраструктуры.

```r
# Формирование параметров инфраструктуры криптобиржи (Формирование данных)
infra_data <- data.frame(
  Segment = c("Trading", "Gateways", "DB"),

  Count_Min = c(6, 3, 3),
  Count_Max = c(10, 5, 5),

  CPU_Cores_Min = c(16, 16, 24),
  CPU_Cores_Max = c(24, 24, 32),

  RAM_GB_Min = c(64, 64, 128),
  RAM_GB_Max = c(128, 128, 256),

  Storage_TB_Min = c(0.5, 0.5, 2),
  Storage_TB_Max = c(1, 1, 4),

  Network_Gbps_Min = c(10, 10, 10),
  Network_Gbps_Max = c(25, 25, 10),

  Price_KUSD_Min = c(15, 12, 18),
  Price_KUSD_Max = c(25, 18, 30),

  Storage_Type = c("Optane", "NVMe", "NVMe"),
  Notes = c("Low latency matching engine",
            "RDMA gateways",
            "Persistent storage / ECC")
)

print(infra_data)
```

```r
# Используем средние значения диапазонов по каждому сегменту

# CPU
cpu_mid <- c(
  Trading = (16 + 24) / 2,
  Gateways = (16 + 24) / 2,
  DB = (24 + 32) / 2
)

seg_cpu_share <- cpu_mid / sum(cpu_mid)

cat("Доли CPU по сегментам:\n")
print(seg_cpu_share)
```

```r
# RAM
ram_mid <- c(
  Trading = (64 + 128) / 2,
  Gateways = (64 + 128) / 2,
  DB = (128 + 256) / 2
)

seg_ram_share <- ram_mid / sum(ram_mid)

cat("\nДоли RAM по сегментам:\n")
print(seg_ram_share)
```

```r
# Storage
storage_mid <- c(
  Trading = (0.5 + 1) / 2,
  Gateways = (0.5 + 1) / 2,
  DB = (2 + 4) / 2
)

seg_storage_share <- storage_mid / sum(storage_mid)

cat("\nДоли Storage по сегментам:\n")
print(seg_storage_share)
```

### **8. Расчет критериев предпочтения для сегментов**

```r
# Trading сегмент
trading_weights <- c(
  CPU = seg_cpu_share["Trading"],
  RAM = seg_ram_share["Trading"],
  Storage = seg_storage_share["Trading"]
)

trading_weights <- trading_weights / sum(trading_weights)

cat("Критерии предпочтения Trading сегмента:\n")
print(trading_weights)
```

```r
# Gateways сегмент
gateways_weights <- c(
  CPU = seg_cpu_share["Gateways"],
  RAM = seg_ram_share["Gateways"],
  Storage = seg_storage_share["Gateways"]
)

gateways_weights <- gateways_weights / sum(gateways_weights)

cat("\nКритерии предпочтения Gateways сегмента:\n")
print(gateways_weights)
```

```r
# DB сегмент
db_weights <- c(
  CPU = seg_cpu_share["DB"],
  RAM = seg_ram_share["DB"],
  Storage = seg_storage_share["DB"]
)

db_weights <- db_weights / sum(db_weights)

cat("\nКритерии предпочтения DB сегмента:\n")
print(db_weights)
```

### **9. Требования к сетевой инфраструктуре (10-25G RDMA)**

```r
# Требования к сети по сегментам
network_requirements <- data.frame(
  Segment = c("Trading", "Gateways", "DB"),
  Network = c("10–25G RDMA", "10–25G RDMA", "10G"),
  Description = c(
    "Минимальная задержка, высокая пропускная способность (matching engine / обработка ордеров)",
    "Маршрутизация/приём трафика, низкая задержка, RDMA",
    "Надёжный доступ к хранилищу и репликациям (10G достаточно)"
  )
)

cat("=== ТРЕБОВАНИЯ К СЕТЕВОЙ ИНФРАСТРУКТУРЕ ===\n\n")
print(network_requirements, row.names = FALSE)
```

```r
# Диапазоны количества узлов по сегментам (из infra_data)
trading_nodes_min <- infra_data$Count_Min[infra_data$Segment == "Trading"]
trading_nodes_max <- infra_data$Count_Max[infra_data$Segment == "Trading"]

gateways_nodes_min <- infra_data$Count_Min[infra_data$Segment == "Gateways"]
gateways_nodes_max <- infra_data$Count_Max[infra_data$Segment == "Gateways"]

db_nodes_min <- infra_data$Count_Min[infra_data$Segment == "DB"]
db_nodes_max <- infra_data$Count_Max[infra_data$Segment == "DB"]

total_nodes_min <- trading_nodes_min + gateways_nodes_min + db_nodes_min
total_nodes_max <- trading_nodes_max + gateways_nodes_max + db_nodes_max

# Оценка количества сетевых портов:
# Берём по 2 сетевых порта на узел (redundancy / bonding)
ports_per_node <- 2

ports_min <- total_nodes_min * ports_per_node
ports_max <- total_nodes_max * ports_per_node

cat("\n=== ОЦЕНКА ПОРТОВ И КОММУТАЦИИ ===\n\n")
cat(sprintf("Узлы всего: %d–%d\n", total_nodes_min, total_nodes_max))
cat(sprintf("Портов (по %d на узел): %d–%d\n", ports_per_node, ports_min, ports_max))

# Примерная оценка количества коммутаторов на 24 порта (верхнеуровневая)
switch_ports <- 24
cat(sprintf("Коммутаторов на %d портов (оценка): %d–%d\n",
            switch_ports,
            ceiling(ports_min / switch_ports),
            ceiling(ports_max / switch_ports)))
```

### **10. Двухуровневая оптимизация**

```r
# Средние значения по диапазонам (вариант 5)

segments <- data.frame(
  Segment = c("Trading", "Gateways", "DB"),

  Count_Min = c(6, 3, 3),
  Count_Max = c(10, 5, 5),

  CPU_Per_Node = c(
    (16 + 24) / 2,
    (16 + 24) / 2,
    (24 + 32) / 2
  ),

  RAM_Per_Node_GB = c(
    (64 + 128) / 2,
    (64 + 128) / 2,
    (128 + 256) / 2
  ),

  Storage_Per_Node_TB = c(
    (0.5 + 1) / 2,
    (0.5 + 1) / 2,
    (2 + 4) / 2
  ),

  Price_Per_Node_K = c(
    (15 + 25) / 2,
    (12 + 18) / 2,
    (18 + 30) / 2
  )
)

print(segments)


# Функция расчета суммарных ресурсов
calc_totals <- function(counts) {

  total_cpu <- sum(counts * segments$CPU_Per_Node)
  total_ram_tb <- sum(counts * segments$RAM_Per_Node_GB) / 1024
  total_storage_tb <- sum(counts * segments$Storage_Per_Node_TB)
  total_price_k <- sum(counts * segments$Price_Per_Node_K)

  return(list(
    CPU = total_cpu,
    RAM_TB = total_ram_tb,
    Storage_TB = total_storage_tb,
    Price_K = total_price_k
  ))
}


# Минимальная конфигурация
min_counts <- segments$Count_Min
min_totals <- calc_totals(min_counts)

# Максимальная конфигурация
max_counts <- segments$Count_Max
max_totals <- calc_totals(max_counts)


cat("=== МИНИМАЛЬНАЯ КОНФИГУРАЦИЯ (HFT) ===\n")
cat(sprintf("Узлов: Trading=%d, Gateways=%d, DB=%d (Всего: %d)\n",
            min_counts[1], min_counts[2], min_counts[3], sum(min_counts)))
cat(sprintf("CPU ядер: %.0f\n", min_totals$CPU))
cat(sprintf("RAM: %.2f ТБ\n", min_totals$RAM_TB))
cat(sprintf("Storage: %.2f ТБ\n", min_totals$Storage_TB))
cat(sprintf("Цена: $%.2fK\n\n", min_totals$Price_K))


cat("=== МАКСИМАЛЬНАЯ КОНФИГУРАЦИЯ (HFT) ===\n")
cat(sprintf("Узлов: Trading=%d, Gateways=%d, DB=%d (Всего: %d)\n",
            max_counts[1], max_counts[2], max_counts[3], sum(max_counts)))
cat(sprintf("CPU ядер: %.0f\n", max_totals$CPU))
cat(sprintf("RAM: %.2f ТБ\n", max_totals$RAM_TB))
cat(sprintf("Storage: %.2f ТБ\n", max_totals$Storage_TB))
cat(sprintf("Цена: $%.2fK\n", max_totals$Price_K))
```
